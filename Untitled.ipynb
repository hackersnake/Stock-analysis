```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader as data
import yfinance as yf

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
import yfinance as yf

# Define the start and end dates
start = '2010-01-01'
end = '2020-12-30'
```

```python
df = yf.download('AAPL', start=start, end=end)
df.head()
```

```python
df.reset_index(inplace=True)

df = df.drop(['Date', 'Adj Close'], axis=1)

df.head()
```

```python
#split data
```

```python
X = df[['Close']]

X_train, X_test = train_test_split(X, test_size=0.3, shuffle=False)  # Set shuffle=False to preserve order

data_training = pd.DataFrame(X_train, columns=['Close'])
data_testing = pd.DataFrame(X_test, columns=['Close'])

print(data_training.shape)
print(data_testing.shape)
```

```python
data_training.head()
```

```python
data_testing.head()
```

```python
scaler=MinMaxScaler(feature_range = (0,1))
```

```python
data_training_array=scaler.fit_transform(data_training)
data_training_array
```

```python
x_trainl = []
y_trainl = []

for i in range(100, len(data_training_array)):
    x_trainl.append(data_training_array[i-50:i, 0]) 
    y_trainl.append(data_training_array[i, 0])  

x_trainl = np.array(x_trainl)  
y_trainl = np.array(y_trainl)  

x_trainl,ytrainl=np.array(x_trainl),np.array(y_trainl)
```

```python
from keras.layers import Dense, Dropout, LSTM
from keras.models import Sequential


```

```python

model = Sequential()

# Add the first LSTM layer
model.add(LSTM(units=50, activation='relu', return_sequences=True, input_shape=(x_trainl.shape[1], 1)))
model.add(Dropout(0.2))

# Add the second LSTM layer
model.add(LSTM(units=60, activation='relu', return_sequences=True))
model.add(Dropout(0.3))

# Add the third LSTM layer
model.add(LSTM(units=80, activation='relu', return_sequences=True))
model.add(Dropout(0.4))

# Add the fourth LSTM layer
model.add(LSTM(units=120, activation='relu'))
model.add(Dropout(0.5))

# Add the output layer
model.add(Dense(units=1))

# Print the model summary
model.summary()
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(x_trainl, y_trainl, epochs=50)
```

```python
model.save('kerass_model.h5')
```
